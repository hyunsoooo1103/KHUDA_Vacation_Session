```python
## Input Box
best_k = None
best_score = 0.0
for k in range(2,9):
  kmeans = KMeans(n_clusters=k, n_init=10, random_state=219).fit(df_scaled)
  df_scaled['cluster'] = kmeans.labels_

  score = silhouette_score(X, df_scaled['cluster'])



silhouette_score(df_scaled.data, df_scaled['cluster'])


answer[1] = best_k
print(answer[1])
```
